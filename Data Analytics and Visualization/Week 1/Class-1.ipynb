Pandas objects: Series and DataFrame

```python
# Series
data = pd.Series([0.25, 0.5, 0.75, 1.0])
print(data)
print(data.values)
print(data.index)
```

```python
# Getting value using index
print(data[1])
print(data[1:4])
print(data[1:3])
```

```python
# Series with custom indices
data = pd.Series([0.25, 0.5, 0.75, 1.0],index=["hello", 5, 3, 7])
print(data)
print(data["hello"])
print(data[7])
print(data["hello":3])
print(data[[5,7]])
```

```python
population_dict = {'Delhi': 19000000,
'Ahmedabad': 5577940,
'Mumbai': 12400000,
'Kolkata': 14900000,
'Bangaluru': 8400000}
data=pd.Series(population_dict)
print(data)
#print(data["Ahmedabad"])
#print(data["Delhi":"Kolkata"])
```

```python
# DataFrame
area_dict = {'Delhi': 1484,
'Ahmedabad': 464,
'Mumbai': 603.4,
'Kolkata': 206.1,
'Bangaluru': 790}

states = pd.DataFrame({'population': population_dict,
                        'area': area_dict})

print(states)
```
